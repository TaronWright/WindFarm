```python
from bs4 import BeautifulSoup
import re
```

```python
#Get the first turbine url of the Windfarm Database
url = df_turbines['Url'][0]
page = requests.get(url)
page.content
soup = BeautifulSoup(page.content)
#Extracts the number of Turbines at the start of the bullet point
number_turbines = soup.find_all("li",{'class':'puce_texte'})[5].next_element[0]
number_turbines
```

```python
#Extracts the url that leads to more information about the turbine
turbine_urls = soup.find_all("a",{'class':'lien_standard'})
for a in turbine_urls:
    if 'turbine' in a['href']:
        turbine_info_url = 'https://www.thewindpower.net/'+a['href']
        print('https://www.thewindpower.net/'+a['href'])
```

```python
turbine_info = requests.get(turbine_info_url)
turbine_soup = BeautifulSoup(turbine_info.content)
rated_power = turbine_soup.find('li', string=re.compile("Rated power*")).next_element
rated_power_column, rated_power_value = rated_power.split(': ')
df['Rated power'] = rated_power_value
```

```python
min_hub_height= turbine_soup.find('li', string=re.compile("Minimum*")).next_element
min_hub_height_column, min_hub_height_value = min_hub_height.split(': ')
df_turbines['Minimum hub height'] = min_hub_height_value 
df_turbines['Minimum hub height']
```

```python
max_hub_height= turbine_soup.find('li', string=re.compile("Maximum*")).next_element
max_hub_height_column, max_hub_height_value = max_hub_height.split(': ')
df_turbines['Maximum hub height'] = max_hub_height_value
df_turbines['Maximum hub height']
```

```python
cutin_speed = turbine_soup.find('li', string=re.compile("Cut-in*")).next_element
cutin_column, cutin_value = cutin_speed.split(': ')
df_turbines['Maximum hub height'] = cutin_value
df_turbines['Maximum hub height']
```

```python
cutoff_speed = turbine_soup.find('li', string=re.compile("Cut-off*")).next_element
cutoff_column, cutoff_value = cutoff_speed.split(': ')
df_turbines['Cut-off wind speed'] = cutoff_value
df_turbines['Cut-off wind speed']
```

```python
rated_speed = turbine_soup.find('li', string=re.compile('Rated wind speed')).next_element
rated_speed_column,rated_speed_value = rated_speed.split(': ')
df_turbines['Rated wind speed'] = rated_speed_value
df_turbines['Rated wind speed']
```

```python
manufacturer = turbine_soup.find('li',string=re.compile('Manufacturer:')).next_element
manufacturer,manufacturer_value = manufacturer.split(': ')
df_turbines['Manufacturer'] = manufacturer_value
df_turbines['Manufacturer']
```

```python
rotor = turbine_soup.find('li',string=re.compile('Rotor*')).next_element
rotor,rotor_value = rotor.split(': ')
df_turbines['Rotor diameter'] = rotor_value
```

```python
df_turbines_original = df_turbines
print(df_turbines_original)
```

```python
df_turbines.dropna(subset=['Url'], inplace=True)
df_turbines
```

```python
#Get the first turbine url of the Windfarm Database
for index in df_turbines.index:
    try:
        page = requests.get(df_turbines['Url'][index])
        print('Base url',df_turbines['Url'][index])
        page.content
        soup = BeautifulSoup(page.content)
        #Extracts all the list elementson the page
        soup.find_all("li")
        #For each list item on the page check if it contains the word 'turbines'
        print('all list elements found')
        for item in soup.find_all("li"):
            item = str(item)
            #if the list text does contain 'turbines' then extract the number
            if 'turbines:' in item:
                pattern = r'(\d+)\s*turbines'
                # Use re.search to find the pattern in the string
                match = re.search(pattern, item)
                print(match)
                if match:
                    print('before assigning number of turbines')
                    df_turbines['Number of Turbines'][index] = match.group(1)
                    print('after assigning number of turbines')
                else:
                    print("Number of turbines not found.")
        turbine_urls = soup.find_all("a",{'class':'lien_standard'})
        for a in turbine_urls:
            if 'turbine' in a['href']:
                turbine_info_url = 'https://www.thewindpower.net/'+a['href']
                print('https://www.thewindpower.net/'+a['href'])
        turbine_info = requests.get(turbine_info_url)
        turbine_soup = BeautifulSoup(turbine_info.content)
        rated_power = turbine_soup.find('li', string=re.compile("Rated power*")).next_element
        rated_power_column, rated_power_value = rated_power.split(': ')
        print("Rated Power:",rated_power_value)
        df_turbines['Rated power'][index] = rated_power_value
        min_hub_height= turbine_soup.find('li', string=re.compile("Minimum hub*")).next_element
        min_hub_height_column, min_hub_height_value = min_hub_height.split(': ')
        df_turbines['Minimum hub height'][index] = min_hub_height_value 
        max_hub_height= turbine_soup.find('li', string=re.compile("Maximum hub*")).next_element
        max_hub_height_column, max_hub_height_value = max_hub_height.split(': ')
        print("Maximum hub height:",max_hub_height_value)
        df_turbines['Maximum hub height'][index] = max_hub_height_value
        cutin_speed = turbine_soup.find('li', string=re.compile("Cut-in*")).next_element
        cutin_column, cutin_value = cutin_speed.split(': ')
        df_turbines['Cut-in wind speed'][index] = cutin_value
        cutoff_speed = turbine_soup.find('li', string=re.compile("Cut-off*")).next_element
        cutoff_column, cutoff_value = cutoff_speed.split(': ')
        df_turbines['Cut-off wind speed'][index] = cutoff_value
        rated_speed = turbine_soup.find('li', string=re.compile('Rated wind speed')).next_element
        rated_speed_column,rated_speed_value = rated_speed.split(': ')
        df_turbines['Rated wind speed'][index] = rated_speed_value
        manufacturer = turbine_soup.find('li',string=re.compile('Manufacturer:')).next_element
        manufacturer,manufacturer_value = manufacturer.split(': ')
        df_turbines['Manufacturer'][index] = manufacturer_value
        rotor = turbine_soup.find('li',string=re.compile('Rotor*')).next_element
        rotor,rotor_value = rotor.split(': ')
        df_turbines['Rotor diameter'][index] = rotor_value
    except:
        print("An error has occured")
```
